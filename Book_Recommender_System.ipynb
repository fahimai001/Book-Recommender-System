# Popularity_Based_Recommendation_System
* We will make a system of top 50 books whose average rating is the highest

```python
ratings_with_name = ratings.merge(books,on='ISBN')
ratings_with_name.head()
```

```python
# We will see all the books correspond to its ratings (Yani har book py kitna number_of_rating howa hai)
num_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].count().reset_index()
num_rating_df.rename(columns={'Book-Rating': 'num_ratings'}, inplace=True)
num_rating_df.head()
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_ratings'}, inplace=True)
avg_rating_df.head()
```

```python
# merge both the dataframe
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
# We will take out all those books 50 books whose avg_rating is greather than or equal to 250
popular_df = popular_df[popular_df['num_ratings']>= 250].sort_values('avg_ratings', ascending=False).head(50)
popular_df.head()
```

```python
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
popular_df.head()
```

```python
# Now ab hum just uun users ko fetch karengay jinho ny minimum 200 ratings kiye hai
x = ratings_with_name.groupby('User-ID')['Book-Rating'].count() > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
filtered_rating.head()
```

```python
y = filtered_rating.groupby('Book-Title')['Book-Rating'].count() >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
print(similarity_score.shape)
```

### Now we will make a function that return 5 suggetions while providing name of the book

```python
def recommend(book_name):
    # fetching index
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Message in a Bottle')
```
